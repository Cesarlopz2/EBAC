```python
# Creamos una nueva columna llamada "Years Playing"

# Importamos el modulo datetime para traer la fecha actual
from datetime import datetime

# Obtenemos la fecha actual.
aÃ±o_actual = datetime.now().year

# Creamos la columna Years Playing
df['Years Playing'] = aÃ±o_actual - df['Joined']

#Comprobamos el resultado
df['Years Playing']




```

```python
# Buscar y mostrar a todos los jugadores de Mexico

jugadores_mexicanos = df[df['Nationality'] == 'Mexico']
jugadores_mexicanos 



```

```python
# Ordenar y mostrar los datos por la columna ReleaseClause (sueldo)
df_releaseclause = df.sort_values(by='Release Clause', ascending=False) #Indicamos que lo queremos en orden descendente
df_releaseclause
```

```python
# Generar un nuevo dataset que contenga el aÃ±o (joined) y el nÃºmero de jugadores (groupby)
players_by_year = df.groupby('Joined').size().reset_index(name='NumberOfPlayers')
players_by_year
```

```python
# Crear Grafico
import seaborn as sns
import matplotlib.pyplot as plt

# Creamos el grafico con la libreria Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=players_by_year, x='Joined', y='NumberOfPlayers', palette='viridis')

# Personalizar el grÃ¡fico con matplotlib.pyplot
plt.title('NÃºmero de Jugadores por AÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('NÃºmero de Jugadores')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
